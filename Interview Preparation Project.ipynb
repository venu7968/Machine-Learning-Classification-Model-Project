* **Decision Tree Classifier:**

* **Random Forest:**

- Similar to Random Forest, it builds multiple models (usually decision trees) on different subsets of the data and averages the results.

* **Gradient Boosting:**
- Another ensemble method that builds models sequentially, where each new model tries to correct the errors of the previous one.
Known for delivering high accuracy, but may take longer to train and is more prone to overfitting without careful tuning.
* **AdaBoost:**
- A boosting method that focuses on the difficult-to-classify examples, giving them higher weight in each new model.
Helps improve model accuracy by focusing on the weaknesses of previous models.
* These models provide different strengths, from simplicity and speed (Decision Trees) to high accuracy and robustness (Random Forest, Gradient Boosting). Ensemble methods like Random Forest, Bagging, and Boosting are particularly effective for this task, as they help reduce overfitting and improve prediction accuracy.

```python
# Fill missing values in the column with the mean value of that column
df.MinTemp = df.MinTemp.fillna(df.MinTemp.mean())
df.MaxTemp = df.MaxTemp.fillna(df.MaxTemp.mean())
df.Rainfall = df.Rainfall.fillna(df.Rainfall.mean())
df.Evaporation = df.Evaporation.fillna(df.Evaporation.mean())
df.Sunshine = df.Sunshine.fillna(df.Sunshine.mean())
df.Humidity9am = df.Humidity9am.fillna(df.Humidity9am.mean())
df.Humidity3pm = df.Humidity3pm.fillna(df.Humidity3pm.mean())
df.Pressure9am = df.Pressure9am.fillna(df.Pressure9am.mean())
df.Pressure3pm = df.Pressure3pm.fillna(df.Pressure3pm.mean())
df.Cloud9am = df.Cloud9am.fillna(df.Cloud9am.mean())
df.Cloud3pm = df.Cloud3pm.fillna(df.Cloud3pm.mean())
df.Temp9am = df.Temp9am.fillna(df.Temp9am.mean())
df.Temp3pm = df.Temp3pm.fillna(df.Temp3pm.mean())

# Fill missing value in the column 'RainToday' with mode of that column
df.RainToday = df.RainToday.fillna(df.RainToday.mode())

# Checking for updated values after filling missing values
df.info()
```

```python
# Convert categorical columns 'RainToday' and 'RainTomorrow' into numerical values
df['RainToday'] = df['RainToday'].map({'Yes': 1, 'No': 0})
df['RainTomorrow'] = df['RainTomorrow'].map({'Yes': 1, 'No': 0})
```

#### Step 3 Performing ML Models

```python
# importing libraries
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier,  AdaBoostClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
```

* **Linear Discrimination Analysis:**  A supervised machine learning algorithm used primarily for classification tasks. It finds a linear combination of features that best separates two or more classes. The main goal of LDA is to project the data onto a lower-dimensional space where the classes are well separated, making it easier to classify new observations. 

```python
# Drop rows with missing values
df_dropped = df.dropna()

# Split the data again into features and target
X_dropped = df_dropped.drop(['RainTomorrow', 'Date',], axis=1)
y_dropped = df_dropped['RainTomorrow']

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_dropped, y_dropped, test_size=0.3, random_state=42)

# Now apply LDA
lda = LinearDiscriminantAnalysis()
lda.fit(X_train, y_train)

# Make predictions
y_pred_lda = lda.predict(X_test)

# Evaluate the LDA model
accuracy_lda = accuracy_score(y_test, y_pred_lda)
cm_lda = confusion_matrix(y_test, y_pred_lda)

# Print results
print(f"LDA Accuracy: {accuracy_lda}")
print(f"LDA Confusion Matrix:\n{cm_lda}")
```

* **KNN:** a non-parametric, supervised learning classifier, which uses proximity to make classifications or predictions about the grouping of an individual data point.

```python
# Drop rows with missing values
df_dropped = df.dropna()

# Split the data again into features and target
X_dropped = df_dropped.drop(['RainTomorrow', 'Date'], axis=1)
y_dropped = df_dropped['RainTomorrow']

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_dropped, y_dropped, test_size=0.3, random_state=42)

# Scale the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Initialize and train KNN model (with k=5)
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train_scaled, y_train)

# Make predictions
y_pred_knn = knn.predict(X_test_scaled)

# Evaluate the KNN model
accuracy_knn = accuracy_score(y_test, y_pred_knn)
cm_knn = confusion_matrix(y_test, y_pred_knn)

# Print results
print(f"KNN Accuracy with Dropped Rows: {accuracy_knn}")
print(f"KNN Confusion Matrix:\n{cm_knn}")
```

* **Gradient Boosting:** is a powerful machine learning technique used primarily for regression and classification tasks. It is an ensemble method that builds models in a sequential manner, where each new model attempts to correct the errors made by the previous models. By combining multiple weak learners (typically decision trees), Gradient Boosting creates a strong predictive model.

```python
# Drop rows with missing values
df_dropped = df.dropna()

# Split the data again into features and target
X_dropped = df_dropped.drop(['RainTomorrow', 'Date',], axis=1)
y_dropped = df_dropped['RainTomorrow']

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_dropped, y_dropped, test_size=0.3, random_state=42)

# Scale the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Initialize and train Gradient Boosting model
gb = GradientBoostingClassifier(random_state=42)
gb.fit(X_train_scaled, y_train)

# Make predictions
y_pred_gb = gb.predict(X_test_scaled)

# Evaluate the Gradient Boosting model
accuracy_gb = accuracy_score(y_test, y_pred_gb)
cm_gb = confusion_matrix(y_test, y_pred_gb)

# Print results
print(f"Gradient Boosting Accuracy with Dropped Rows: {accuracy_gb}")
print(f"Gradient Boosting Confusion Matrix:\n{cm_gb}")
```

* **AdaBoost:** is a popular ensemble learning algorithm primarily used for classification tasks. It combines the predictions of several weak learners (typically decision trees) to produce a strong predictive model. The main idea behind AdaBoost is to focus on the training examples that are difficult to classify by previous models, thereby improving overall accuracy.

```python
# Initialize and train AdaBoost model
ada = AdaBoostClassifier(n_estimators=100, random_state=42)
ada.fit(X_train_scaled, y_train)

# Make predictions
y_pred_ada = ada.predict(X_test_scaled)

# Evaluate the AdaBoost model
accuracy_ada = accuracy_score(y_test, y_pred_ada)
cm_ada = confusion_matrix(y_test, y_pred_ada)

# Print results
print(f"AdaBoost Accuracy with Dropped Rows: {accuracy_ada}")
print(f"AdaBoost Confusion Matrix:\n{cm_ada}")
```
